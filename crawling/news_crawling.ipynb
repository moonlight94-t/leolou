```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# Selenium으로 웹 드라이버를 실행
service = Service(executable_path=ChromeDriverManager().install()) #service설정하지 않으면 웹드라이버가 위치한 경로를 전달해줘야함.
driver= webdriver.Chrome(service=service)

# url 전달
url='https://news.kbs.co.kr/news/pc/category/category.do?ref=pSiteMap#20240913&1'
driver.get(url)
# 페이지 로딩 기다려달라는 값 전달, 최대10초 빨리 열리면 더 안기다림 
wait=WebDriverWait(driver, 10)

html= driver.page_source
driver.quit()  # 실행되고 있는 드라이버를 꺼주어야함
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
box_contents=soup.find_all('a',class_="box-content flex-style")
box_contents # 데이터가 담겨있음 
```

```python
# 페이지네이션 : 페이지 구조보고 for문 돌리는데 매번 크롬 드라이버를 여닫을 것인지
# 아니면 url을 바꿔줄 것인지 
```

```python
url_list=[]
title_list=[]
body_list=[]
date_list=[]
service = Service(executable_path=ChromeDriverManager().install())

for i in range(1,5):
    driver= webdriver.Chrome(service=service)

    url=f'https://news.kbs.co.kr/news/pc/category/category.do?ref=pSiteMap#20240913&{i}'
    driver.get(url)
 
    wait=WebDriverWait(driver, 10)

    html= driver.page_source
    driver.quit()
    soup = BeautifulSoup(html, 'html.parser')
    box_contents=soup.find_all('a',class_="box-content flex-style")
    for box_content in box_contents:
        url="http://news.kbs.co.kr" + box_content['href']
        title= box_content.find('p', class_='title').text
        body = box_content.find('p', class_='news-text').text
        date = box_content.find('span', class_='date').text
    
        url_list.append(url)
        title_list.append(title)
        body_list.append(body)
        date_list.append(date)
```
