```python
url_list=[]
title_list=[]
body_list=[]

urls=['https://news.kbs.co.kr/news/pc/view/view.do?ncd=8058606','https://news.kbs.co.kr/news/pc/view/view.do?ncd=8059116']
for url in urls:
    html_doc=requests.get(url).text
    soup=BeautifulSoup(html_doc, 'html.parser')
    title=soup.find('h4', class_='headline-title').text
    body=soup.find('div', class_='detail-body font-size').text
    url_list.append(url)
    title_list.append(title)
    body_list.append(body)

data12={'뉴스url':url_list, '제목':title_list,'내용':body_list}
df12=pd.DataFrame(data12)
```
