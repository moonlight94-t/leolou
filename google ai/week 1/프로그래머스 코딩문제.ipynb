```python
file_names = ["새파일1.txt", "새파일2.txt", "새파일3.txt"]
fruits = [
    ["apple", "banana", "cherry"],
    ["grape", "orange", "pear"],
    ["kiwi", "peach", "plum"]
]


for file_name, fruit_list in zip(file_names, fruits):
    with open(file_name, 'w', encoding='utf-8') as f:
        for fruit in fruit_list:
            f.write(fruit + "\n")

f.close()
```
