```python
df.dropna(how='any',inplace=True) # 결측치 있는 행 제거 , 제거한다고 데이터의 의미가 크게 달라지나? 데이터가 없는건 비교의 의미가 없으니까 논외
```

```python
df.tail() #!4:06
```

```python
#평창군 한해동안 가장 방문객이 많았던 해
df['연 방문객']=df.loc[:,'1월':'12월'].sum(axis=1)

total={}
for i in range(2010,2022):  # 결측치 삭제하면서 2022년은 NaN
  total[df.loc[df['연도']==i]['연 방문객'].sum(axis=0)]=i

print(f'{total[max(total)]}년에 {int(max(total))}명')

# indexing sum 사용법
```

```python
#값 갱신
#i=df.columns[4,16]
#df.loc['i']

before=df.loc[:,'1월'].sum(axis=0)
most=1
for i in df.columns[4:16]:
  now=(df.loc[:,i].sum(axis=0))
  if before < now:
    before=now
    most=i

print(f'{most}월에 {before}')
```

```python
# 평창군에서 가장 핫플레이스는 어디인가? 최근 5년동안 평균 방문객이 가장 많은곳
df2=df.loc[df['연도']>2016].copy()

for i in set(df.loc[df['연도']>2016]['관광지']):
  now=df2.loc[df2['관광지']=='국립두타산자연휴양림'].loc[:,'1월':'12월'].sum(axis=0).sum(axis=0)
  if before == None :
    before=now
  elif before < now:
    before=now
    most=i

print(f'{i}장소에 5년간 {now}명방문')
```

```python
# 관광지가 추가되거나 삭제된 곳은 어디인가?
result=[]
for i in range(2010,2022):
  now=set(df.loc[df['연도']==i]['관광지'])
  if before == None :
    before=now

  if len(before)>len(now):
    result.extend(list(before-now))
  elif len(now)>len(before):
    result.extend(list(now-before))


print(f'{len(result)}개가 사라지거나 추가\n목록 : {result}')
```
