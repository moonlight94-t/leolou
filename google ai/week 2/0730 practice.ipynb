```python
import random

random.randint(1,6)

#던지는 횟수를 늘려야 1/6 확률에 근사, 수렴함
#카멜표기법 diceValue 변수이름 지정 잘해주는것도 중요
iter=20
ret2=[0 for _ in range(6)]
for _ in range(iter):
  ret=random.randint(1,6)
  ret2[ret-1]+=1

ret3= list(map(lambda x : x/iter,ret2))

print(ret3)
```

```python
#*args 튜플 **kwarg 딕셔너리 여러개의 입력값 받는 함수
#꼭 순서대로 안넣더라도 인위적으로 a,b 값을 지정해줄 수 있다.
def addnsub(a,b):
  return a+b , a-b

#인수, 결과 튜플로 앞부분

def add_many(*args):
  result=0
  for i in args:
    result+=i
  return result
```

```python
def createlog():
  import datetime
  now = datetime.datetime.now()
  format_date =now.strftime('%Y-%m-%d')
  global name
  name=format_date+'_log.txt'
  f=open(name,'w')
  f.write('날짜,시간,메뉴,가격\n')
  f.close()
  return
```

```python
def accounting():
  import os
  import time
  file_path = '/content/'+name  # 오늘 날짜 입력
  file_stat = os.stat(file_path)
  modified_timestamp = file_stat.st_mtime
  modified_date = time.strftime('%Y-%m-%d,%H:%M:%S', time.localtime(modified_timestamp))
  log=f'{modified_date}'+','+f'{calcul(order_number)}'
  for i in order_number:
    log+=','+f'{menu_dic2[int(i)]}'
  log+='\n'
  f=open(name,'a')
  f.write(log)
  f.close()
  return
```
