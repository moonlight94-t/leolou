```python
class Car:
  make = ''
  model=''
  year =0

  def __init__(self, make, model, year):
    self.make=make
    self.model=model
    self.year=year

  def __str__(self):
    return f'This car is an {self.model} produced in {self.year} by {self.make}'

car1=Car('hyundai', 'avante', 2020)

print(car1)
```

```python
#super 상속

class Truck(Car):
  payload_capacity =0

  def __init__(self, make, model, year, payload_capacity):
    super().__init__(make,model,year)
    self.payload_capacity=payload_capacity

  def load_cargo(self, weight):
    if weight <= self.payload_capacity:
      print(f'Loaded cargo of {weight} kg succesfully')
    else:
      print('denied')

truck1 = Truck("ford",'f150',2022,2000)
print(truck1)
truck1.load_cargo(1500)
```

```python
#이런식으로 list값 빼올 수 있음 unpack
def maxd(point1, point2):
  x1, y1 = point1
  x2, y2 = point2

  maxd= max(math.fabs(x2-x1), math.fabs(y2-y1))
  return maxd

maxd= maxd([1,2],[4,6])

print(maxd)

x1, y1=[2,3]
print(f'{x1}{y1}')
```

```python
def area(a,b,c):
  s=(a+b+c)/2

  area=math.sqrt(s*(s-a)*(s-b)*(s-c))
  return area

area=area(3,4,5)

print(area)
```

```python
def shuffle_deck():
  suits=['하트', '다이아몬드', '스페이드','클로버']
  ranks=['2','3','4','5','6','7','8','9','10','j','k','q','k','a']
  deck=[[rank,suit] for suit in suits for rank in ranks]
  rd.shuffle(deck)
  return deck

print (shuffle_deck())
```

```python
def frequent(s,k):
  counts=Counter(s)
  sorted_counts=sorted(counts.items(),key=lambda x:x[1], reverse=True)
  topk =sorted_counts[:k]
  #return [char for char in topk]  #[('c', 7), ('a', 4)]
  #return [char for char, _ in topk] #['c', 'a'] comprehension for문에서 _ 사용
  return [topk[i][0] for i in range(len(topk))]

st2='aaaabbbcccccccd'
k=2

print(frequent(st2,k))
```
