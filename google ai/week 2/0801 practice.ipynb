```python
def name_generator(name_number):
  Last=['김','이','박','최','정','강','조','윤','장','임']
  First=['민준', '서준','도윤', '예준',	'시우' ,'하준' ,'지호', '주원', '지후' ,'준우' ,'서윤' ,'서연' ,'지우' ,'하윤' ,'서현' ,'하은' ,'민서' ,'지유' '윤서']
  First1t=set()
  First2t=set()
  name=[]
  Lastweight=[10689959,7306828,4192074,2333927,2151879,1176847,1055567,1020547,992721,823921]
  Last2=rd.choices(Last,weights=weight(Lastweight),k=name_number)

  for i in First:
      First1t.add(i[0])
      First2t.add(i[1])
      First1=list(First1t)
      First2=list(First2t)
  while len(name) < name_number:
    a=rd.choice(First1)
    b=rd.choice(First2)
    while a==b:
      a=rd.choice(First1)
      b=rd.choice(First2)
    name.append(Last2[len(name)]+a+b)
    name=set(name)
    name=list(name)
  return name
```

```python
def id_generator(name_number,id_leng):
  id_string='ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789'
  id=[]
  for _ in range(name_number):
    id.append(''.join(rd.choices(id_string,k=id_leng)))
  return id
```

```python
def softmax(X):            #지수함수형태로 양수를 만들고 확률개념으로 바꾸어줌 / 지수함수 특성상 큰 값에 가중치 큼 / 이상치에 약함
 exp_a = np.exp(X)
 sum_exp_a = np.sum(exp_a)
 y = exp_a / sum_exp_a
 return y
```

```python
# 나이생성기
def age_generator(name_number):
  age_set=list(np.arange(13,81))
  age= rd.choices(age_set,weights=np.random.normal(1,1,len(age_set)),k=name_number) #weights=np.random.normal(0,1,len(age_set)),
  return age
```

```python
def phone_number(name_number):
  number_string='0123456789'
  phone=[]
  for _ in range (name_number):
    phone1=rd.choices(number_string,k=4)
    phone2=rd.choices(number_string,k=4)
    phone.append('010-'+''.join(phone1)+'-'+''.join(phone2))
  return phone
```

```python
def visit_place(name_number):
  place=['강남','판교','도봉','성북']
  a= rd.choices([0,1],weights=[5/7,2/7],k=name_number)
  ans=[]
  for i in range(name_number):
    if a[i]==0:
      b=rd.choices(place,weights=[0.4,0.4,0.1,0.1],k=1)
      ans.append((rd.choices(place,weights=[0.4,0.4,0.1,0.1],k=1))[0])
    else:
      ans.append((rd.choices(place,weights=[0.2,0.2,0.3,0.3],k=1))[0])
  return ans      #가중치
```

```python
def save_csv(name_number, id_leng):
  data='이름, ID, 나이, 방문지점, 전화번호\n'
  for i in range(name_number):
    data+= name_generator(name_number)[i]+', '+id_generator(name_number,id_leng)[i]+', '+str(age_generator(name_number)[i])+', '+visit_place(name_number)[i]+', '+phone_number(name_number)[i]+'\n'
  f=open('data_set.csv','w')
  f.write(data)
  f.close()
  return
```
