```python
#database 에 덮어지는 경우 데이터베이스 복원이 힘드므로 주의할 것
from pathlib import Path
import os
import random
n=0
while n < 30:
  letter_len=6
  letters_set='abcdefghijklmnopqrstuvwxyz'
  random_list = random.sample(letters_set, letter_len)
  name=''.join(random_list)+'.txt'
  file_path = Path(name)
  file_path.touch()
  n+=1

cur_list3=os.listdir('.')
print(cur_list3)

cur_list4=list(filter(lambda x : os.path.splitext(x)[1]=='.txt',cur_list3))
cur_list5=list(filter(lambda x : x[0]=='a', cur_list4))
#list끼리 차집합 구하기

for i in cur_list5:
  os.remove(i)

#os.remove([i for cur_list5]) #comprehension 축약을 어디에 사용할 것인지 list를 만드는 방법

cur_list6=list(filter(lambda x : os.path.splitext(x)[1]=='.txt',os.listdir()))
print(cur_list6)

for i,j in enumerate(cur_list6) :
  changename = "myText"+str(i+1).zfill(2)+'.txt' #01 02 03 04 / 메서드 사용 후 새로 값 받아서 해야하는 것과 그대로 넣어도 되는 것의 차이 sort
  os.rename (j,changename)

ans=os.listdir()
ans.sort()

print(ans)

#.endswith startswith
```

```python
def solution(number):
	count = 0
	for i in range(1, number + 1):
		current = i
		while current != 0:
			if current%10==3 or current%10==6 or current%10==9 :
				count += 1
			current = current // 10
	return count


# //10의 활용 한자리씩 따로 확인
```
